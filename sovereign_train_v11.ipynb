```python
# 5. Visualize Results (Static Plot)
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os

# Find the latest CSV results file
files = glob.glob("outputs/sovereign_v5/*.csv")
if files:
    latest_file = max(files, key=os.path.getctime)
    print(f"📈 Plotting results from: {latest_file}")
    df = pd.read_csv(latest_file)
    
    plt.figure(figsize=(10, 5))
    plt.plot(df.index, df['reward'], label='Cumulative Reward', color='#00d4aa')
    plt.title('Sovereign Agent: GRPO Reward Convergence')
    plt.xlabel('Batch Index')
    plt.ylabel('Reward Value')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()
else:
    print("❌ No CSV results found yet. Make sure training finished successfully!")
```
